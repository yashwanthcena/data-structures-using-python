```python
#Creating the desired Tree
class Tree(object):
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data)
def dfs(root):
    if root:
        print(root.data)
        dfs(root.left)
        dfs(root.right)
root=Tree('A')
root.left=Tree('B')
root.left.left=Tree('S')
root.left.right=Tree('R')
root.left.left.left=Tree('W')
root.left.left.left.left=Tree('Z')
root.left.right.left=Tree('R')
root.left.right.left=Tree('Y')
root.left.right.right=Tree('A')
root.right=Tree('U')
root.right.left=Tree('E')
root.right.right=Tree('P')
root.right.right.left=Tree('G')
root.right.right.right=Tree('J')
print('Postorder:')
print(postorder(root))
print('--------------------------------------')
print('dfs:')
print(dfs(root))
```
